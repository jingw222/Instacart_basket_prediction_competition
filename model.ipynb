```python
import tensorflow as tf
from functools import partial
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
```

```python
tf.reset_default_graph()

n_instances, n_inputs = X_train.shape
n_hidden1 = 10
n_hidden2 = 5
n_outputs = 2

initial_learning_rate = 0.01
decay_steps = 1000
decay_rate = 1/10
global_step = tf.Variable(0, trainable=False)
learning_rate = tf.train.exponential_decay(initial_learning_rate, global_step,
                                           decay_steps, decay_rate)
# dropout_rate = 0.01

X = tf.placeholder(tf.float32, shape=(None, n_inputs), name='X')
y = tf.placeholder(tf.int64, shape=(None), name='y')
is_training = tf.placeholder(dtype=tf.bool, shape=(), name='is_trainging')

he_init = tf.contrib.layers.variance_scaling_initializer()

my_batch_norm_layer = partial(tf.layers.batch_normalization, 
                              training=is_training, 
                              momentum=0.9)
my_dense_layer = partial(tf.layers.dense, 
                         kernel_initializer=he_init)
# my_dropout_layer = partial(tf.layers.dropout, 
#                            rate=dropout_rate, 
#                            training=is_training, 
#                            seed=13114)

with tf.name_scope('dnn'):
    
    hidden1 = my_dense_layer(X, n_hidden1, name='hidden1')
    bn1 = tf.nn.elu(my_batch_norm_layer(hidden1), name='bn1')
#     bn1_dropped = my_dropout_layer(bn1)
    
    hidden2 = my_dense_layer(bn1, n_hidden2, name='hidden2')
    bn2 = tf.nn.elu(my_batch_norm_layer(hidden2), name='bn2')
#     bn2_dropped = my_dropout_layer(bn2)
    
    logits = my_dense_layer(bn2, n_outputs, activation=tf.nn.sigmoid, name="logits")
    extra_update_op = tf.get_collection(tf.GraphKeys.UPDATE_OPS)
    
with tf.name_scope('loss'):
    xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)
    loss = tf.reduce_mean(xentropy, name='loss')
    
with tf.name_scope("train"):
    optimizer = tf.train.AdamOptimizer(learning_rate)
    train_op = optimizer.minimize(loss, global_step=global_step)    

with tf.name_scope("eval"):

    correct = tf.nn.in_top_k(logits, y, 1)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

pred = tf.argmax(logits, name='predicted', axis=1)
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
n_epochs = 2
batch_size = 500
n_batches = n_instances // batch_size

n_batches
```

```python
with tf.Session() as sess:
    init.run()

    kf = KFold(3, shuffle=True)
    for i, (train_idx, valid_idx) in enumerate(kf.split(X_train)):
        X_training, X_valid, y_training, y_valid = X_train[train_idx], X_train[valid_idx], y_train[train_idx], y_train[valid_idx]
        scaler = StandardScaler()
        X_training = scaler.fit_transform(X_training)
        X_valid = scaler.transform(X_valid)
        for epoch in range(n_epochs):
            shuffled_idx = np.random.permutation(X_training.shape[0])
            for j, batch_index in enumerate(np.array_split(shuffled_idx, n_batches)):
                X_batch, y_batch = X_training[batch_index], y_training[batch_index]
                sess.run([train_op, extra_update_op], feed_dict={is_training: True, X: X_batch, y: y_batch})
                
                if j % 200 == 0:
                    loss_training, acc_training = sess.run([loss, accuracy], feed_dict={is_training: False, X: X_training, y: y_training})
                    loss_valid, acc_valid = sess.run([loss, accuracy], feed_dict={is_training: False, X: X_valid, y: y_valid})    
                    print('Fold:', i, 'epoch:', epoch, 'batch:', j)
                    print('training loss:', loss_training, 'validation loss:', loss_valid)
                    print('training accuracy:', acc_training, 'validation accuracy:', acc_valid)
            save_path = saver.save(sess, './ckpt/my_model.ckpt', global_step=epoch)
    predicted_train = pred.eval(feed_dict={is_training: False, X: X_train})
#     predicted_pred = pred.eval(feed_dict={is_training: False, X: X_pred})
        
print('Training finished!')
```
